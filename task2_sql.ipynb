```python
import pandas as pd

df = pd.read_csv("../data/cleaned_superstore.csv")

df.head()
```

```python
import sqlite3

conn = sqlite3.connect("superstore.db")

print("Database Created Successfully")
```

```python
df.to_sql(
    "sales",
    conn,
    if_exists="replace",
    index=False
)

print("Data Imported Successfully")
```

```python
query = "SELECT * FROM sales LIMIT 5"

pd.read_sql(query, conn)
```

```python
query = """
SELECT Category,
SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Category
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT [Customer Name]
FROM sales
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM sales
WHERE Sales > 500
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT [Product Name], Sales
FROM sales
ORDER BY Sales DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT Category,
COUNT(*) AS Total_Orders
FROM sales
GROUP BY Category
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT Category,
SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Category
HAVING SUM(Sales) > 100000
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT [Product Name],
SUM(Sales) AS Total_Sales
FROM sales
GROUP BY [Product Name]
ORDER BY Total_Sales DESC
LIMIT 5
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT Region,
SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Region
ORDER BY Total_Sales DESC
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT [Customer Name],
SUM(Sales) AS Total_Sales
FROM sales
GROUP BY [Customer Name]
ORDER BY Total_Sales DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT Category,
SUM(Profit) AS Total_Profit
FROM sales
GROUP BY Category
ORDER BY Total_Profit DESC
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT Category,
AVG(Sales) AS Average_Sales
FROM sales
GROUP BY Category
"""

pd.read_sql(query, conn)
```
